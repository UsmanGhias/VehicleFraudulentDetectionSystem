## Introduction:

- Vehicle Fraudulent Detection System is a sophisticated solution designed to detect and prevent fraudulent activities in the context of vehicle-related insurance claims. With the increasing prevalence of fraudulent claims, it has become imperative for insurance companies to implement robust systems that can accurately identify suspicious activities and mitigate financial losses. 
- This system utilizes advanced machine learning algorithms and data analysis techniques to analyze various attributes and patterns associated with vehicle claims, aiming to distinguish between genuine and fraudulent claims. Our vehicle fraudulent detection system plays a vital role in promoting fair and trustworthy insurance practices within the industry.

## 1. Business Objective
- Minimize Financial Losses: The primary objective of the Vehicle Fraudulent Detection System is to minimize financial losses incurred by the company due to fraudulent activities. 
- Protect Company Reputation: Another key objective is to safeguard the company's reputation and brand image. By effectively identifying and combating fraudulent activities, the system helps maintain trust and confidence among customers, stakeholders, and the general public.
- Enhance Operational Efficiency: The Vehicle Fraudulent Detection System seeks to enhance operational efficiency by automating the fraud detection process.

### Dataset Description

- **Month:** The month in which the accident occurred.
- **WeekOfMonth:** The week of the month in which the accident occurred.
- **DayOfWeek:** The day of the week on which the accident occurred.
- **Make:** The make or brand of the vehicle involved in the accident.
- **AccidentArea:** The location or area where the accident took place.
- **DayOfWeekClaimed:** The day of the week on which the claim was made.
- **MonthClaimed:** The month in which the claim was made.
- **WeekOfMonthClaimed:** The week of the month in which the claim was made.
- **Sex:** The gender of the policyholder.
- **MaritalStatus:** The marital status of the policyholder.
- **Age:** The age of the policyholder.
- **Fault:** Indicates whether the policyholder was at fault in the accident.
- **PolicyType:** The type of insurance policy.
- **VehicleCategory:** The category of the vehicle (e.g., sedan, SUV, truck).
- **VehiclePrice:** The price range or value of the vehicle.
- **PolicyNumber:** The unique policy number associated with the claim.
- **RepNumber:** The representative number handling the claim.
- **Deductible:** The amount of deductible for the insurance policy.
- **DriverRating:** The rating of the policyholder as a driver.
- **Days:Policy-Accident:** The number of days between the policy start date and the accident.
- **Days:Policy-Claim:** The number of days between the policy start date and the claim date.
- **PastNumberOfClaims:** The number of previous claims made by the policyholder.
- **AgeOfVehicle:** The age of the vehicle.
- **AgeOfPolicyHolder:** The age group of the policyholder.
- **PoliceReportFiled:** Indicates whether a police report was filed for the accident.
- **WitnessPresent:** Indicates whether a witness was present at the accident.
- **AgentType:** The type of insurance agent involved.
- **NumberOfSuppliments:** The number of supplementary claims filed.
- **AddressChange-Claim:** Indicates whether the policyholder changed address after the claim.
- **NumberOfCars:** The number of cars insured under the policy.
- **Year:** The year in which the accident occurred.
- **BasePolicy:** The base insurance policy type.
- **FraudFound:** Indicates whether fraud was detected (target variable).

### Feature Engineering Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler, RobustScaler
from sklearn.impute import SimpleImputer

# Preprocessing steps for categorical features
categorical_transformer = Pipeline(steps=[
    ('cat_imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore'))
])

# Preprocessing steps for numerical features
numerical_transformer = Pipeline(steps=[
    ('num_imputer', SimpleImputer(strategy='mean')),
    ('scaler', RobustScaler())
])
```

```python
# Combine the transformers
preprocessor = ColumnTransformer(transformers=[
    ('categorical', categorical_transformer, cat_columns),
    ('numerical', numerical_transformer, num_columns)
])

# Create the feature engineering pipeline
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
])
```

**Before Feature Engineering**

```python
X_train.head()
```

```python
X_train.shape
```

**After Feature Engineering**

```python
pipeline.fit_transform(X_train)
```

```python
pipeline.fit_transform(X_train).shape
```

```python
X_train_scaled = pipeline.fit_transform(X_train)
X_test_scaled = pipeline.transform(X_test)
```

**Dump Feature Engineering Pipeline**

```python
import joblib
```

```python
joblib.dump(pipeline, 'models/pipeline.pkl')
```

```python
# Initializing classifiers
classifiers = [
    LogisticRegression(),
    DecisionTreeClassifier(),
    SVC(),
    KNeighborsClassifier(),
    GaussianNB(),
    RandomForestClassifier()
]
```

```python
# Lists to store training and testing errors
train_errors = []
test_errors = []

# Lists to store classification metric scores
accuracy_scores = []
precision_scores = []
recall_scores = []
f1_scores = []
roc_auc_scores = []

```

```python
# Training and evaluating each classifier
for classifier in classifiers:
    classifier.fit(X_train_scaled, y_train)
    train_pred = classifier.predict(X_train_scaled)
    test_pred = classifier.predict(X_test_scaled)
    
    train_errors.append(1 - accuracy_score(y_train, train_pred))
    test_errors.append(1 - accuracy_score(y_test, test_pred))
    
    accuracy_scores.append(accuracy_score(y_test, test_pred))
    precision_scores.append(precision_score(y_test, test_pred))
    recall_scores.append(recall_score(y_test, test_pred))
    f1_scores.append(f1_score(y_test, test_pred))
    roc_auc_scores.append(roc_auc_score(y_test, test_pred))
```

```python
classifiers_name = ['Logistic Regression', 'Decision Tree', 'SVC', 'KNN', 'Naive Bayes', 'Navie Bayes', 'Random Forest']
scores = accuracy_scores

plt.figure(figsize=(10, 6))
x = np.arange(len(classifiers_name))
width = 0.35

plt.bar(x, scores, width)
plt.xlabel('Classifier')
plt.ylabel('Accuracy Score')
plt.xticks(x, classifiers_name, rotation=45)
plt.title('Accuracy Scores for Different Classifiers')
plt.show()
```

## Choose Best Model

```python
# Finding the best classifier based on the chosen metric
best_classifier_index = np.argmax(accuracy_scores)
best_classifier_name = type(classifiers[best_classifier_index]).__name__
print("Best Classifier (based on Accuracy-Score):", best_classifier_name)

```

```python
# Printing confusion matrix of the best classifier
best_classifier = classifiers[best_classifier_index]
best_classifier.fit(X_train_scaled, y_train)
best_pred = best_classifier.predict(X_test_scaled)
confusion_mat = confusion_matrix(y_test, best_pred)
print("\nConfusion Matrix:")
confusion_mat
```
